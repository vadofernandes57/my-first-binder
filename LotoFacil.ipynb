```python
balls_15 = ['ball_1',  'ball_2', 'ball_3', 'ball_4', 'ball_5', 'ball_6', 'ball_7', 
            'ball_8', 'ball_9', 'ball_10', 'ball_11', 'ball_12', 'ball_13', 'ball_14', 'ball_15']
#for train
train['mean_15_balls'] = train[balls_15].mean(axis=1)
train['var_15_balls'] = train[balls_15].var(axis=1)
train['sum__15_balls'] = train[balls_15].sum(axis=1)

#for test
test['mean_15_balls'] = test[balls_15].mean(axis=1)
test['var_15_balls'] = test[balls_15].var(axis=1)
test['sum__15_balls'] = test[balls_15].sum(axis=1)
```

```python
X_train = []
X_train_columns = train.columns
c = 0
for v in train.values:
  info = v[:15]
  binary = v[15:]
  index = [k for k, i in enumerate(binary) if i==1]
  for i in index:
    c+=1
    for k in range(len(binary)):
      if k == i:
        binary_transformed = list(copy.copy(binary))
        binary_transformed[i] = 0
        X_train.append(list(info) + binary_transformed + [X_train_columns[15+k]] + [c])


X_train = pd.DataFrame(X_train)
X_train.columns = ['ID', 'Date', 'mean_15_balls', 'var_15_balls', 'sum__15_balls', 'ball_1',  'ball_2', 'ball_3', 'ball_4', 'ball_5', 'ball_6', 'ball_7', 
                 'ball_8', 'ball_9', 'ball_10', 'ball_11', 'ball_12', 'ball_13', 'ball_14', 'ball_15', '1b', '2b', '3b', '4b', '5b', '6b', '7b', '8b', '9b', '10b', 
                 '11b', '12b', '13b', '14b', '15b', 'product_pred', 'ID_2']
```

```python
# Modeling

X = X_train.drop(columns = ['ID', 'ID_2', 'Date', 'product_pred'], axis=1)
y = y_train



from mlxtend.classifier import EnsembleVoteClassifier
import warnings
import sklearn
import lightgbm as lgb
from xgboost import XGBClassifier, plot_importance

from catboost import CatBoostClassifier
warnings.filterwarnings('ignore')

# Create boosting classifier
lgbm = lgb.LGBMClassifier(boosting_type='gbdt', objetive='multiclass',
                         learning_rate=0.017, num_iterations=950, #950best 0.017
                         bagging_fraction=0.85, feature_fraction=0.9, reg_lambda=1.3, reg_alpha=1.91, num_class=42, bagging_freq = 9, 
                         colsaple_bytree=0.75, n_jobs= -1, seed=236)

cat = CatBoostClassifier(n_estimators=1700, task_type = "GPU", learning_rate = 0.022, depth = 7, random_state = 236, #Learining_rate
                        )
xgb_boost = XGBClassifier(max_depth = 5, objective = 'multi:softprob', n_estimators = 1700, learning_rate = 0.02, #max_depth=5 Learning
                          num_class = 25, tree_method = 'gpu_hist', seed = 236)
boost_array = [lgbm,cat,xgb_boost]

eclf = EnsembleVoteClassifier(clfs=[lgbm, cat, xgb_boost], voting='soft') #soft
```
